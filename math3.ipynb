### Задание 3
Построить: окружность, эллипс, гиперболу

```python
x = np.linspace(-5,5,1000)
```

```python
R = 5
y1 = np.sqrt(R**2 - x**2)
y2 = - np.sqrt(R**2 - x**2)

fig, ax = plt.subplots()
fig.set_size_inches(5, 5)

ax.plot(x,y1, 'green')
ax.plot(x,y2, 'green')

plt.title('Окружность')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
a = 2
b = 5
y3 = np.sqrt(a**2*(1 - x**2/b**2))
y4 = -np.sqrt(a**2*(1 - x**2/b**2))

plt.plot(x,y3, 'green')
plt.plot(x,y4, 'green')

plt.title('Эллипс')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
a = 2
b = 5
for i,value in enumerate(x):
    if value > -a and value < a:
        x[i] = None

y5 = np.sqrt(b**2*(x**2/a**2 - 1))
y6 = -np.sqrt(b**2*(x**2/a**2 - 1))

plt.plot(x,y5, 'green')
plt.plot(x,y6, 'green')

plt.title('Гипербола')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

### Задание 4

Допустим у нас есть 2 точки с координатами (x1,y1,z1) и (x2,y2,z2), которые принадлежат плоскости, заданной уравнением Ax+By+Cz+D = 0. Тогда появляется система уравнений:

1)$$A*x + B*y + C*z + D = 0$$
2)$$A*x_1 + B*y_1 + C*z_1 + D = 0$$
3)$$A*x_2 + B*y_2 + C*z_2 + D = 0$$

Если из перого уравнения последовательно вычесть второе и третье, то получим:

4)$$A*(x - x_1) + B*(y - y_1) + C*(z - z_1) = 0$$
5)$$A*(x - x_2) + B*(y - y_2) + C*(z - z_2) = 0$$

Если из 4го уравнения вычесть 5е, то получим уравнение 6). Плюс у нас есть уравнение прямой:

6)$$A*(x_2 - x_1) + B*(y_2 - y_1) + C*(z_2 - z_1) = 0$$
7)$$\frac{x - x_1}{x_2 - x_1} = \frac{y - y_1}{y_2 - y_1} = \frac{z - z_1}{z_2 - z_1} = m$$

Если уравнения 6) и 7) решать как систему уравнений и получится выйти на исходное уравнение плоскости, значит такая прямая ей принадлежит. Перемножим их:

7)$$A*(x_2 - x_1) * m + B*(y_2 - y_1) * m + C*(z_2 - z_1) * m = 0$$
$$A*(x_2 - x_1) * \frac{x - x_1}{x_2 - x_1} + B*(y_2 - y_1) * \frac{y - y_1}{y_2 - y_1} + C*(z_2 - z_1) * \frac{z - z_1}{z_2 - z_1} = 0$$
$$A*(x - x_1) + B*(y - y_1) + C*(z - z_1) = 0$$

Последнее уравнение принадлежит плоскости $$A*x + B*y + C*z + D = 0$$ Значит прямая тоже ей принадлежит.
